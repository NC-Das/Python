```python
num = int(input())

number = input()

n = number.split()

odd =0
neg =0
pos =0
even = 0


for i in range(num):
    if int(n[i])%2==0:
        even= even + 1
    else:
        odd= odd +1
    if int(n[i])<0:
        neg = neg +1
    if int(n[i])>0:
        pos = pos +1
    
print("Even:", even,"\nOdd:", odd,"\nPositive:", pos,"\nNegative:", neg)
```
