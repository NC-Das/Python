```python
num = int(input())

number = num
sum = 0

while num>0:
    sum = sum*10 + num%10
    num//=10

if sum == number:
    print("Mirror Number")
else:
    print("Not Mirror Number")
```

```python
word_count = {}
with open("data.txt", 'r') as file:
            for line in file:
                words = ''.join(
                    char.lower() if char.isalnum() or char.isspace() else ' '
                    for char in line
                ).split()
                for word in words:
                    word_count[word] = word_count.get(word, 0) + 1
            max_word = max(word_count, key= word_count.get)
            print(max_word)
file.close()
```
