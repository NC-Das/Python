```python
numbers = input()
num = numbers.split()
a = int(num[0])
b = int(num[1])

try:
    res = (a**2 + b**2)/(a-b)
except ZeroDivisionError:
    print("Error: Division not possible")
else:
    print(res)

    
```

```python
num = int(input())

number = num
sum = 0

while num>0:
    sum = sum*10 + num%10
    num//=10

if sum == number:
    print("Mirror Number")
else:
    print("Not Mirror Number")
```
